```kotlin
import javax.xml.crypto.Data

val df = DataFrame.readCSV("/Users/esteban/Downloads/mental.csv")
df.count()
```

```kotlin
val gender_avg = df
    .groupBy { Gender }
    .aggregate {
        mean("Age") into "Age"
    }
    .sortBy {Age}
gender_avg
```

```kotlin
val brazil = df.filter { Country == "Brazil" }
brazil
```

```kotlin
brazil.add("Happy + 1 "){`Happiness Score`+  1}
```

```kotlin
val brazil_avg_sleep = brazil.groupBy("Country").aggregate {
    max(`Sleep Hours`) into "Max Sleep Hours"
    min(`Sleep Hours`) into "Min Sleep Hours"
    mean(`Sleep Hours`) into "Average Sleep Hours"
}
brazil_avg_sleep
```


# Average sleep hours by country

```kotlin
val avg_sleep = df.groupBy {Country}.aggregate {
    mean(skipNA = true) {`Sleep Hours`} into "avg sleep hours"
}.sortByDesc("avg sleep hours")
avg_sleep
```
