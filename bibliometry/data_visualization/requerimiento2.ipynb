 # Clasificiación
### Autores principales con más apariciones

```kotlin
//Mapa que almacena la frecuencia de aparición de cada autor
val auth_freq = mutableMapOf<String, Int>()

val first_authors = df.get { Author }.toList()

//Cuenta la frecuencia de aparición de cada primer autor en el set de articulos.
first_authors.forEach { author ->
    auth_freq[author] = auth_freq[author]?.plus(1) ?: 1
}

auth_freq.toList().sortedByDescending { (_, value) -> value }.take(15)
```

## Autores con más apariciones

```kotlin
val all_auths = df.get { Author }.toList()

all_auths

val all_auths_freq = mutableMapOf<String, Int>()

all_auths.forEach { author ->
    all_auths_freq[author] = all_auths_freq[author]?.plus(1) ?: 1
}

all_auths_freq.toList().sortedByDescending { (_, value) -> value }
```

## Densisdad de recursos por año


```kotlin
val densidad_anio = df.groupBy { it.`Publication year (YYYY-MM-DD)` }.aggregate {
    count() into "Cantidad"
}.sortByDesc("Cantidad")

densidad_anio
```

## Densidad de recurso por año y tipo de recurso

```kotlin

val densidad_anio_tipo = df.groupBy {
    `Type of reference` and `Publication year (YYYY-MM-DD)`
}.aggregate {
    count() into "Cantidad"
}
densidad_anio_tipo
```

## Densidad por tipo de recurso

```kotlin
val densidad_tipo = df.groupBy { `Type of reference` }.aggregate { count() into "Cantidad" }
densidad_tipo
```

## Journals con más apariciones

```kotlin
val journals_freq = df.groupBy {it.`Journal name`}.aggregate {
    it.get("Journal name") into "Journal"
    count() into "Cantidad"
}.sortByDesc("Cantidad")

journals_freq
//Journal/Periodical name: full format.
//Journal/Periodical name: full format.1
```

# 15 journals con más apariciones

```kotlin
journals_freq.take(15)
```

# Publishers con más apariciones

```kotlin
val publishers_freq = df.groupBy { it.`Publisher` }.aggregate {
    count() into "Cantidad"
}.sortByDesc("Cantidad")

publishers_freq
```
