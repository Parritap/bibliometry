 # Clasificiación
### Autores principales con más apariciones

```kotlin
//Mapa que almacena la frecuencia de aparición de cada autor
val auth_freq = mutableMapOf<String, Int>()

val first_authors = df.get { authors }.toList().map {
  it.split(",").first()
      }

//Cuenta la frecuencia de aparición de cada primer autor en el set de articulos.
first_authors.forEach { author ->
    auth_freq[author] = auth_freq[author] ?: 0 + 1
}

auth_freq.toList().sortedByDescending { (_, value) -> value }.take(15)
```

## Autores con más apariciones

```kotlin
val all_auths = df.get { authors }.toList().map {
   it.split(",")
}.flatten()

all_auths

val all_auths_freq = mutableMapOf<String, Int>()

all_auths.forEach { author ->
    all_auths_freq[author] = all_auths_freq[author] ?: 0 + 1
}

all_auths_freq.toList().sortedByDescending { (_, value) -> value }
```

## Densisdad de recursos por año


```kotlin
val densidad_anio = df.groupBy { year }.aggregate {
    count() into "Cantidad"
}.sortByDesc("Cantidad")

densidad_anio
```

## Densisdad de recurso por año y tipo de recurso

```kotlin
val densidad_anio_tipo = df.groupBy { type and year }.aggregate { count() into "Cantidad" }
densidad_anio_tipo
```

## Densidad por tipo de recurso

```kotlin
val densidad_tipo = df.groupBy { type }.aggregate { count() into "Cantidad" }
densidad_tipo
```
